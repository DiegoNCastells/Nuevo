```python
import csv,sqlalchemy
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim 
```

```python
arch='query_result_2022-09-23T18 17 38.503901Z.csv'
```

```python
data=pd.read_csv(arch,encoding='utf-8')
```

```python
data=data.iloc[:101,:]
data
```

```python
cp_=data['zip']
city_=data['city']
state_=data['state']
street_=data['street']
```

```python
print(f'la cantidad de elemntos es :',len(cp_))
listaDirections=[]
for i in range(len(cp_)):
    try:
        #cp_[i]+','+
        #print(state_[i]+','+city_[i]+','+street_[i])
                
        #listaDirections.append(street_[i]+' '+city_[i]+' '+state_[i]+' '+cp_[i]+' United State')
        listaDirections.append(cp_[i]+' '+city_[i]+' '+state_[i]+' UNITED STATES')

    except:
        listaDirections.append('error')
```

```python
for direction in listaDirections:
    print(direction)
```

```python
import time
```

```python
geolocator = Nominatim(user_agent="DiegoPila")
listaCordenates=[]
for direction in listaDirections:
    try:
        if direction!=0:
            location=geolocator.geocode(direction)
            listaCordenates.append([location.latitude, location.longitude])
        else:
            print('hay un error')
        #print(location)
        print(location.latitude, location.longitude)
    except:
        print('aviso')
        listaCordenates.append(['',''])
```
