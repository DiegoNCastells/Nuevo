```python
def Brand_M(marcas,auto):
    try:
            dato=auto
            for x in marcas:
                if( str(x[0]) in auto ):
                    auto = auto.replace(x[0],'')
                    modelo=auto.split()[0] 
                    marca = x[0]
                    motor=re.findall(r"[-+]?(?:\d*\.\d+|\d+)",auto)
                    auto=auto.split(str(motor[0]))  
                    motor=motor[1]
                    line=auto[1].split()
                    #hasta aca anda bien
                    #print(line)
                    linea=""
                    for i in range(1,len(line)):
                        linea=linea+' '+line[i]
                    #print(linea)
                    #print(linea[1])
                    return (marca,modelo,motor,linea)
    except:
        print(dato)
        return (None,None,None,None)
```
