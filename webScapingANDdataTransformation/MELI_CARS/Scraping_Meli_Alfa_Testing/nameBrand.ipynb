```python
import requests
```

```python
aux='https://www.diariomotor.com/marcas/'
response = requests.get(aux)
```

```python
from bs4 import BeautifulSoup
```

```python
if response.status_code==200 :
    soup = BeautifulSoup(response.text,'html.parser')
    lista = soup.find('div',attrs={'css-xt4val e17elydr3'})
```

```python
names= lista.find_all('span',attrs={'css-1ulzmi e1rmgh9t1'})
```

```python
import sqlite3 as sql
```

```python
con=sql.connect('brandCars.db')
table=con.cursor()
```

```python
#table.execute('''CREATE TABLE brandCars(IdCar integer,nameBrand text)''')
```

```python
listNames=[]
idNames=[]
i=0
for name in names:
    idNames.append(i)
    listNames.append(name.get_text())
    i+=1
db=zip(idNames,listNames)
```

```python
table.executemany('INSERT INTO brandCars VALUES(?,?)',db)
```

```python
con.commit()
```

```python
for row in table.execute('SELECT * FROM brandCars'):
    print(row)
```
